Scraping website tidak membutuhkan authentikasi. cara scraping step by step bisa dilihat disini : https://nbviewer.org/github/arofiqimaulana/Scrab-Table-on-Website/blob/master/Scrab%20Premier%20League%20Table.ipynb

```python
URL = "https://www.bbc.com/sport/football/premier-league/table"
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
page = requests.get(URL)
soup = BeautifulSoup(page.content, 'html.parser')
tables = soup.find_all("table")
```

```python
table = tables[0]
tab_data = []

for row in table.find_all("tr"):
    element = [cell.text for cell in row.find_all(["th","td"])]
    tab_data.append(element)

df = pd.DataFrame(tab_data)
df.head()
```

```python
df_clean = df.iloc[1:,:]
```
