```python
import pandas as pd
```

## 1. Import Data

```python
df_data = pd.read_excel('MilitaryExpenditure2.xls',sheet_name='Data')
df_countries = pd.read_excel('MilitaryExpenditure2.xls',sheet_name='Metadata - Countries')
df_indicator = pd.read_excel('MilitaryExpenditure2.xls',sheet_name='Metadata - Indicators')
```

```python
df_data
```

## 2. Transformasi Data

```python
# Menggunakan melt untuk mengubah kolom tahun menjadi baris
df_melted = df_data.melt(
    id_vars=['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code'], 
    var_name='Year', 
    value_name='Value'
)
```

```python
df_melted
```

## 3. Cleaning Data

### 3.1 Handle Missing Values

```python
# handle missing values
df_melted.fillna('',inplace=True)
```

### 3.2 Mengambill data yang tidak kosong

```python
# filter yang tidak kosong
df_melted2 = df_melted[df_melted.Value!='']
df_melted2
```

### 3.3 Mengubah tipe data

```python
# mengubah tipe data
```

```python
df_melted2.info()
```

```python
df_melted2['Value'] = df_melted2['Value'].astype(float) 
```

```python
df_melted2.info()
```

### 3.4 Filter beberapa negara

```python
# filter hanya Indonesia dan Belgium
country = ['Indonesia','Belgium']
df_melted3 = df_melted2[df_melted2['Country Name'].isin(country)]
df_melted3
```

```python
# tipe  varibel Value sudah menjadi float64 yang awalnya object  
df_melted3.info()
```

### 3.5 Membuat kolom baru

```python
def category_exp(v):
    if v > 1:
        return 'big'
    else:
        return 'small'
```

```python
df_melted2['category'] = df_melted2['Value'].apply(category_exp)
```

```python
df_melted2
```
